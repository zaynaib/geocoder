```python
def fullDirections(direction):
    if direction == 'W':
        return 'WEST'
    if direction == 'E':
        return 'EAST'
    if direction == 'S':
        return 'SOUTH'
    if direction == 'N':
        return 'NORTH'
    else:
        return direction
    
```

```python
omec_clean_locations['St_PreDir']= omec_clean_locations['directional'].apply(fullDirections)
omec_clean_locations.head()
```
